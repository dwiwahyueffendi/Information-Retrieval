```python
import csv
import xlsxwriter
from tqdm import tqdm
```

```python

# Inisiasi variabel
i=0
wordList1 = [] #tempat simpan kata dan jumlahnya yg sudah di proses dari dokumen/label 1
wordCheck1 = [] #tempat simpan kata baru dari dokumen 1 dan tempat untuk cek jika kata sudah pernah terbaca (khusus dokumen 1)
wordList2 = [] #tempat simpan kata dan jumlahnya yg sudah di proses dari dokumen/label 2
wordCheck2 = []#tempat simpan kata baru dari dokumen 2 dan tempat untuk cek jika kata sudah pernah terbaca (khusus dokumen 1)
wordCount = [] #tempat simpan sementara kata dan jumlahnya 

temp = ["","","","","",""] #tempat simpan kata secara bertahap
punctuations ='''!()-[]{};:'"\,<>./?@#$%^&*_~''' #tempat simpan tanda baca 
insSuf = ["ing","ed","pre","ly","ful"] #tempat simpan imbuhan

cleanWord = "" #tempat kata diproses


#fungsi untuk menurutkan berdasarkan nilai terbanak ke tersedikit
def Sort(sub_li): 
  
    # reverse = None (Sorts in Ascending order) 
    # key is set to sort using second element of  
    # sublist lambda has been used 
    return(sorted(sub_li, key = lambda x: x[1],reverse = True))     


with open('trimText.txt','r', encoding='utf-8') as file: #membuka file trimText.txt
   
    # reading each line     
    for line in tqdm(file): #membaca tiap baris pada file (tqdm untuk memunculkan proses)
       
        if("__label__1" in line): #membaca jika baris memiliki kata "__label__1"
            # reading each word         
            for word in line.split(' '): #membaca tiap kata pada baris
                #membaca tiap huruf bada baris
                for evrAlph in word: 
                    if (evrAlph not in punctuations and evrAlph not in "\n"): #jika huruf bukan termasuk tanda baca 
                        cleanWord = cleanWord + evrAlph #maka huruf akan disusun satu persatu hungga membentuk kata
                                                        #tanpa tanda baca
                
                #tempat memproses tiap kata sehingga tidak ada imbuhan
                for fix in insSuf:
                    if fix in cleanWord:

                        temp[0] = cleanWord.replace("ing","") #jika kata terdapat ing maka ing dihilangkan

                        temp[1] = temp[0].replace("ed","") #jika kata terdapat ed maka ing dihilangkan

                        temp[2] = temp[1].replace("pre","") #jika kata terdapat pre maka ing dihilangkan

                        temp[3] = temp[2].replace("ly","") #jika kata terdapat ly maka ing dihilangkan
                        
                        temp[4] = temp[3].replace("ful","") #jika kata terdapatful maka ing dihilangkan
                        
                        cleanWord = temp[3] #mengganti nilai kata sebelum diproses menjadi kata setelah diproses
              
            #mengecek tiap kata agar tidak ada kata yang sama pada list/daftar
                            
                if(cleanWord not in wordCheck1):
                    print(cleanWord)
                    wordCount.append(cleanWord)
                    wordCount.append(1)
                    wordList1.append(wordCount)
                    wordCheck1.append(cleanWord)
                    wordCount = []
                else:
                    #menambahkan nilai frekuensi jika ada kata yang sama
                    for m in wordList1:
                        if m[0].casefold() == cleanWord.casefold():
                            m[1]+=1
            #menghapus nilai pada variabel agar dapat digunakan kembali pada pengulangan berikutnya
                cleanWord = ""
                   
    #proses yang sama terjadi pada dokumen 2
        if("__label__2" in line):
            # reading each word         
            for word in line.split(' '):
                
                for evrAlph in word:
                        if evrAlph not in punctuations and evrAlph not in "\n":
                            cleanWord = cleanWord + evrAlph
                            
                
                for fix in insSuf:
                    if fix in cleanWord:

                        temp[0] = cleanWord.replace("ing","")

                        temp[1] = temp[0].replace("ed","")

                        temp[2] = temp[1].replace("pre","")

                        temp[3] = temp[2].replace("ly","")
                        
                        temp[4] = temp[3].replace("ful","")
                        
                        cleanWord = temp[3]
                
                if(cleanWord not in wordCheck2):
                #Storing Word
                    wordCount.append(cleanWord)
                    wordCount.append(1)
                    wordList2.append(wordCount)
                    wordCheck2.append(cleanWord)
                    wordCount = []
                else:
                    for n in wordList2:
                        if n[0].casefold() == cleanWord.casefold():
                            n[1]+=1
                            
                cleanWord = ""

#mengurutkan nilai dari terbesar ke terkecil
wordList2 = Sort(wordList2)
wordList1 = Sort(wordList1)


#membuka dan membuat file excel hello.xlsx, jika sudah ada akan di replace
workbook = xlsxwriter.Workbook('hello.xlsx')
worksheet = workbook.add_worksheet()

index = 2

#nilai pada daftar beserta nilai akan dicetak pada file excel secara berurutan 

for x in wordList1:
    worksheet.write("A"+str(index),x[0])#kolom A akan ditempati oleh kata pada dokumen 1 
    worksheet.write("B"+str(index),x[1])# kolom B akan ditempati oleh nilai pada dokumen 1
    index+=1


index = 2

for x in wordList2:
    worksheet.write("C"+str(index),x[0])#kolom c akan ditempati oleh kata pada dokumen 2 
    worksheet.write("D"+str(index),x[1])# kolom D akan ditempati oleh nilai pada dokumen 2
    index+=1

            
 #menutup file excel           
workbook.close()

          # displaying the words       
```
