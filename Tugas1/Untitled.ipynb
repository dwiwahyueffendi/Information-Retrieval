```python
for i in range(20) :
    if nilai[i] == 5:
        result = review[i].find('great')
        if result != -1 :
            cari=review[i].split(' ')
            ambil=collections.Counter(cari).most_common()
            
            for kata in ambil:
              ##  print("%s: %d"%(kata[0],kata[1]))
                if kata[0] == 'great':
                   print("%s\t: %d"%(kata[0],kata[1]))
```
